```python
from PIL import Image
import numpy as np
import os
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
import cv2 
import random
import skimage.measure
```

### Original Image 1 (image1.png)

```python
image = "sampleImages/image1"
ext = ".png"
img1 = cv2.imread(image + ext,1) 

plt.figure(figsize=(14,6))
histogram_blue = cv2.calcHist([img1],[0],None,[256],[0,256])
plt.plot(histogram_blue, color='blue') 
histogram_green = cv2.calcHist([img1],[1],None,[256],[0,256]) 
plt.plot(histogram_green, color='green') 
histogram_red = cv2.calcHist([img1],[2],None,[256],[0,256]) 
plt.plot(histogram_red, color='red')
plt.xlabel('Pixel values', fontsize=16)
plt.ylabel('Pixel count', fontsize=16) 
plt.show()
```

```python
org_entropy_1 = skimage.measure.shannon_entropy(img1)
print(org_entropy_1)
```
